This blog post describes an analysis of correlation coefficient between urban rate and the income per person based on GapMinder dataset that is available at the course website. 

```python
%matplotlib inline
import pandas 
import numpy as np
import scipy
import matplotlib.pyplot as plt
import pip    
def install(package):
   pip.main(['install', package])

install('seaborn')

import seaborn
```

```python
data = pandas.read_csv('gapminder.csv', low_memory=False)
```

```python
data['incomeperperson'] = pandas.to_numeric(data['incomeperperson'], errors='coerce')
data['urbanrate']  = pandas.to_numeric(data['urbanrate'], errors='coerce')
```

```python
scat1 = seaborn.regplot(x="urbanrate", y="incomeperperson", fit_reg=True, data=data)
plt.xlabel('urbanrate')
plt.ylabel('incomeperperson')
```

```python
data_clean=data.dropna()

print ('association between urban rate and income per person:')
r = scipy.stats.pearsonr(data_clean['urbanrate'], data_clean['incomeperperson'])
print (r)
print ('r-squared:')
print (r[0]**2)
```

These results indicate a significant moderate positive linear relationship between urban rate and the income per person. Furthermore about 24% of the income per person variance can be explained/predicted by the urban rate.
